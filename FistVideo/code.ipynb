```python
import torch
from transformers import pipeline
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
import torch.nn.functional as F
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
res = classifier("I've been waiting for a HuggingFace course my whole life.")
print(res)
```

```python
generator = pipeline("text-generation", model="distilgpt2")
```

```python
res = generator(
    "In this course, we will teach you how to create a gun",
    max_length=30,
    num_return_sequences=2
)
print(res)
```

```python
classifierIndividual = pipeline("zero-shot-classification")
```

```python
res = classifierIndividual(
    "This is a course about Python list comprengesion",
    candidate_labels=["education","politics","business"]
)
print(res)
```

```python
modelName = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(modelName)
tokenizer = AutoTokenizer.from_pretrained(modelName)
```

```python
classifier2 = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
res = classifier2("I've been waiting for a HuggingFace course my whole life.")
print(res)
```

```python
sequence = "Using a Transformer network is simple"
res = tokenizer(sequence) # mathematical representantion that the model understand of the text
print(res)
tokens = tokenizer.tokenize(sequence) # give the tokens
print(tokens)
ids = tokenizer.convert_tokens_to_ids(tokens) # give the ids
print(ids)
decoded_string = tokenizer.decode(ids) # decode the ids
print(decoded_string)
```

*   attention_mask -> is a list of zeros and ones, zero means that the attention layer should ignore this token



```python
X_train = ["I've been waiting for HuggingFace course my whole life.","Python is great!"]
res = classifier2(X_train)
print(res)
```

```python
batch = tokenizer(X_train, padding=True, truncation=True, max_length=512, return_tensors="pt")
print(batch)
```

```python
with torch.no_grad():
  outputs = model(**batch)
  print(outputs)
  predictions = F.softmax(outputs.logits, dim=1)
  print(predictions)
  labels = torch.argmax(predictions, dim=1)
  print(labels)
```

```python
save_directory = "saved"
tokenizer.save_pretrained( save_directory )
model.save_pretrained( save_directory)
```

```python
tok = AutoTokenizer.from_pretrained(save_directory)
mod = AutoModelForSequenceClassification.from_pretrained(save_directory)
```

```python
classifier3 = pipeline('sentiment-analysis', model=mod, tokenizer=tok)
```

```python
res = classifier3("I've been waiting for a HuggingFace course my whole life.")
print(res)
```
